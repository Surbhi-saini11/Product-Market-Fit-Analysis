Q6. List the top two most reported disease illnesses in each state over the past three 
years, along with the corresponding average Air Quality Index (AQI) for that 
period. 
---

```python
import pandas as pd
data=pd.read_csv('day-wise-state-wise-air-quality-index-aqi-of-major-cities-and-towns-in-india.csv')
health_data = pd.read_csv('state_wise_disease.csv', encoding_errors='replace')
health_data.drop(['unit','note'],axis=1,inplace=True)
```

```python
health_data=health_data[health_data['year']>=2022]
illnesses=health_data.groupby(['state','disease_illness_name'])['cases'].sum().reset_index(name='Total_cases')
illnesses=pd.DataFrame(illnesses)
res=[]
for state in illnesses['state'].unique():
    df=illnesses[illnesses['state']==state].sort_values(by='Total_cases',ascending=False)
    top_2=df.head(2)
    res.append(top_2)
res=pd.concat(res)

# avg AQI by states
state_aqi_filter=data[data['date']>='2022-01-01']
state_aqi_avg=state_aqi_filter.groupby('state')['aqi_value'].mean().reset_index(name='avg_aqi')
state_aqi_avg

new=pd.merge(res[['state','disease_illness_name']],state_aqi_avg[['state','avg_aqi']],on='state')
new = new.reset_index(drop=True)
new
```

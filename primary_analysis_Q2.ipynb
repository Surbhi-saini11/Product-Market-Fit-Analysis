2. List out top 2 and bottom 2 prominent pollutants for each state of southern India. 
(Consider data post covid: 2022 onwards)
---

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns

data=pd.read_csv('day-wise-state-wise-air-quality-index-aqi-of-major-cities-and-towns-in-india.csv')
```

```python
# filtering dataset for the post covid analysis (2022 onwards)
df=data[data['date']> '31-12-2021']

# filter data further for the analysis of prominent pollutants for each state of southern India
southern_states=['Kerala','Karnataka','Telangana','Tamil Nadu','Andhra Pradesh']

df=df[df['state'].isin(southern_states)]
df['prominent_pollutants']=df['prominent_pollutants'].str.split(',')
df=df.explode('prominent_pollutants')
res=df.groupby(['state','prominent_pollutants']).size().reset_index(name='frequency')
res=pd.DataFrame(res)


result=[]
for state in res['state'].unique():
    subset=res[res['state'] == state].sort_values(by='frequency',ascending=True)
    top_2=subset.head(2)
    bottom_2=subset.tail(2)
    result.append(top_2)
    result.append(bottom_2)

print(pd.concat(result))
```

```python
#average AQI in each southern state
avg_aqi=df.groupby('state')['aqi_value'].mean().reset_index(name='avg_aqi')
avg_aqi
```

#### Key Insights :

South India does not a uniform air quality problem - it has 5 distinct pollution profiles requiring 5 distinct product and marketing strategies. A one-size-fits-all air purifiers will underserve every market. AirPure's competitive advantage lies in city-specific product configurations driven by this localised pollution data.
